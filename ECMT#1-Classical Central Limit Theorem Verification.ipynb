```python
%matplotlib inline
# import matplotlib
import matplotlib.pyplot as plt
# for latex equations
# from IPython.display import Math, Latex
# # for displaying images
# from IPython.core.display import Image
# import seaborn
import seaborn as sns
# settings for seaborn plotting style
sns.set(color_codes=True)
# settings for seaborn plot sizes
sns.set(rc={'figure.figsize':(5,5)})
from random import sample
import statistics
import math
```

```python
# define function to calculate average of a list of numbers
def average(list):
    return sum(list)/len(list)

# This is the function to output average sample value given certain sample size and iterations
def avg_sample_frequency (data,number_of_samples,iterations):
    global sample_distribution_of_mean
    sample_distribution_of_mean = []
    i = 0
    list = data.tolist()
    while i < iterations :
        i = i + 1
        sample_distribution_of_mean.append(average(sample(list,number_of_samples)))
    
# This will define the number of samples to take from the population and the number of times average is calculated  
number_of_samples =100;
number_of_iterations =1000;
```

```python
# import uniform distribution
from scipy.stats import uniform

# random numbers from uniform distribution
n = 10000
start = 0
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)
```

```python
ax = sns.distplot(data_uniform,
                  bins=100,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 0.5,'alpha':1,'color':'blue'})
print("mean of poulation is " + str(average(data_uniform)))
print("standard error given the number of the sample is " + str(statistics.stdev(data_uniform)/math.sqrt(number_of_samples)))
```

```python
# Distribution of sample means for uniform distribution
avg_sample_frequency(data_uniform,number_of_samples,number_of_iterations)
ax = sns.distplot(sample_distribution_of_mean,
                  bins=100,
                  kde=False,
                  color='skyblue',
                  hist_kws={"linewidth": 0.5,'alpha':1,'color':'blue'})

print("mean of the sample average is " + str(average(sample_distribution_of_mean)))
print("standard deviation of the sample average  is " + str(statistics.stdev(sample_distribution_of_mean)))
```
